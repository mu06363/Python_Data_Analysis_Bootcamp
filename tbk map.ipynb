```python
import geopy
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent='South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    try: crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}
    except : crd = {"lat": 'error', "lng": 'error'}
    return crd

lat_list = []
lng_list = []

for temp in df['address']:
    lat_list.append(geocoding(temp)['lat'])
    lng_list.append(geocoding(temp)['lng'])

df['lat'] = lat_list
df['lng'] = lng_list
```

```python
import folium
from folium.map import Popup

lat = 37.566660
lng = 126.978778

center = [lat, lng]

seoul_map = folium.Map(location=center, zoom_start=12, tiles='cartodbpositron')

import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul_map)


for temp in range(len(df['name'])):
    try:
        name = df.iloc[temp]['name']
        lat = df.iloc[temp]['lat']
        lng = df.iloc[temp]['lng']
        folium.Marker(location=[lat, lng], icon=folium.Icon(color='red'), popup=name).add_to(seoul_map)
    except : pass

seoul_map
```

```python
# 더본코리아 지역별 매장 수
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)

tbk_map = folium.Map(location=[37.559819, 126.963895], zoom_start=12, tiles='cartodbpositron')

folium.GeoJson(seoul_geo, name='지역구').add_to(tbk_map)

tbk_map.choropleth(geo_data=seoul_geo,
                   data=gu_data_tbk, 
                   fill_color='Blues',
                   fill_opacity=0.5,
                   line_opacity=0.2,
                   key_on='properties.name',
                   legend_name="지역구별 더본코리아 업종 수"
                   )

tbk_map
```
