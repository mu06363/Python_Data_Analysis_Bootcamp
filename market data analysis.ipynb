```python
df['음식점'] = np.where(df['서비스_업종_코드_명'] == '한식음식점', '식',
                     np.where(df['서비스_업종_코드_명'] == '호프-간이주점', '식',
                              np.where(df['서비스_업종_코드_명'] == '커피-음료', '식',
                                       np.where(df['서비스_업종_코드_명'] == '육류판매', '식',
                                                np.where(df['서비스_업종_코드_명'] == '치킨전문점', '식',
                                                         np.where(df['서비스_업종_코드_명'] == '패스트푸드점', '식',
                                                                  np.where(df['서비스_업종_코드_명'] == '일식음식점', '식',
                                                                            np.where(df['서비스_업종_코드_명'] == '양식음식점', '식',
                                                                                     np.where(df['서비스_업종_코드_명'] == '중식음식점', '식', 'X')))))))))
```
