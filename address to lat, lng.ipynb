```python
import geopy
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent='South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    try: crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}
    except : crd = {"lat": 'error', "lng": 'error'}
    return crd

lat_list = []
lng_list = []

for temp in df['address']:
    lat_list.append(geocoding(temp)['lat'])
    lng_list.append(geocoding(temp)['lng'])

df['lat'] = lat_list
df['lng'] = lng_list

df.head()
```
